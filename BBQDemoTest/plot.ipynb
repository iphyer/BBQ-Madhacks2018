```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
truth = np.loadtxt(filename, delimiter=',' )
fig = plt.figure()
ax = fig.add_subplot(111, aspect='equal')
im = np.array(Image.open('../rawData/set3/200kV_500kx_p2nm_8cmCL_grain1_0090 - Copy.jpg'), dtype=np.uint8)
ax.imshow(im.astype(np.uint8), cmap='gray')
recList = list()

color = {
    0 : "red",
    1 : "blue",
    2 : "yellow"
}

# Adding blue rectangle from predict
for rect in truth:
    recList.append(
        patches.Rectangle(
            (rect[1], rect[2]),
             np.abs(rect[3] - rect[1]),
             np.abs(rect[4] - rect[2]),
             fill=False,
             edgecolor = color.get(rect[0]),
             linewidth=2
        )
    )

# # Adding red rectangle for truth
# for rect in truth:
#     recList.append(
#         patches.Rectangle(
#             (rect[0], rect[3]),
#              np.abs(rect[2] - rect[0]),
#              np.abs(rect[3] - rect[1]),
#              fill=False,
#              edgecolor = "red"
#         )
#     )

# plot the graph
for p in recList:
    ax.add_patch(p)

plt.plot()
plt.show()
fig.savefig('rect.png', dpi=90, bbox_inches='tight')
```
